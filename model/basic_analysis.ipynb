```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import pylab 
import scipy.stats as stats
import input_data
```

```python
train_data = input_data.get_data(date='20160817')
num_of_cols = train_data.shape[1] - 1
```

```python
x_train = train_data.iloc[:,0:num_of_cols]
y_train = train_data.iloc[:,num_of_cols]
```
